```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

url = "https://raw.githubusercontent.com/Anuj020/CsvFiles/main/fifa21_male2.csv"

fifa = pd.read_csv(url,low_memory=False)
fifa.head()
```

```python
# Distribution of overall player ratings

plt.figure(figsize=(10, 6))
sns.histplot(fifa['Age'], kde=True)
plt.title('Distribution of Overall Ratings')
plt.xlabel('Overall Rating')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
x = fifa['Age']
y = fifa['Stamina']
plt.bar(x,y)
plt.title('Player Age vs. Stamina')
plt.xlabel('Age')
plt.ylabel('Stamina')
plt.show()
```

```python
# Bar plot of preferred foot
plt.figure(figsize=(6, 4))
foot_counts = fifa['foot'].value_counts()
plt.bar(foot_counts.index, foot_counts.values, color=['lightgreen', 'lightblue'])
plt.title("Preferred Foot of Players")
plt.xlabel("Preferred Foot")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
top_nationalities = fifa['Nationality'].value_counts().head(5).index
df_top_nationalities = fifa[fifa['Nationality'].isin(top_nationalities)]
plt.boxplot([df_top_nationalities[df_top_nationalities['Nationality'] == nationality]['OVA'] for nationality in top_nationalities],
            labels=top_nationalities, showfliers=False)
plt.title("Player Ratings by Top 5 Nationalities")
plt.xlabel("Nationality")
plt.ylabel("Overall Rating")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
avg_age_by_nationality = fifa.groupby('Nationality')['Age'].mean().sort_values(ascending=False).head(10)
plt.bar(avg_age_by_nationality.index, avg_age_by_nationality.values, color='lightblue')
plt.title("Average Age of FIFA 21 Players by Nationality (Top 10)")
plt.xlabel("Nationality")
plt.ylabel("Average Age")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(fifa['Age'], fifa['POT'], color='darkgreen', alpha=0.5)
plt.title("Age vs. Potential")
plt.xlabel("Age")
plt.ylabel("Potential Rating")
plt.show()
```
